```python
def convertData(filename):
    import pandas as pd
    import numpy as np

    print('Importing Excel')
    excel_file = pd.ExcelFile(filename)
    print('Importing Done')
    df_total = pd.DataFrame()
    
    sheets = excel_file.sheet_names

    for sheet in sheets:
        print(sheet)
        df_temp = excel_file.parse(sheet_name = sheet, header=None)
        df_total = df_total.append(df_temp)
        
    
    print("Fetching data from all sheets")
    
    data = df_total.copy()
    data.columns = ['name', 'todrop']
    data = data.drop(['todrop'], axis = 1)
    
    print("Dropped useless second column")

    name = data['name'].str.contains("NAME   = ")

    end = data['name'].str.find("END")

    
    print("Fetching Start and End Index of each Block")
    
    temp = name.tolist()
    count = 0 
    name_index = []
    
    for i in temp:
        count = count + 1
        if  i is True:
            name_index.append(count)

    
    temp = end.tolist()
    count = 0 
    end_index = []
    
    
    for i in temp:
        count = count + 1
        if  i == 0:
            end_index.append(count)

    print("Fetching Done")

    print("Converting to list of dictionaries")

    final_list = []
    
    if len(name_index) == len(end_index):
        
        for i in range(len(name_index)):
            yolo = []      
            yolo = data['name'][name_index[i]-1:end_index[i]-1].tolist()
            dictionary = {}

            for y in yolo:
                if y.find('=') != -1:
                    dictionary[y.split("=")[0].strip()] = y.split("=")[1].strip()

            final_list.append(dictionary)

    print("Convertion done")
    
    print("Converting list of dic to DataFrame")
    df_1 = pd.DataFrame(final_list)
    
    print("Function done!")
    
    return df_1

    
```

```python
convertData('Data/u20.xlsx')
```
