```python
import pandas as pd
file = open("dataICCPRT.txt", "r")
print(file)
```

```python
templist = []
for line in file:
    a = line.strip()
    b = a.split()
    templist.append(b)
```

```python
df = pd.DataFrame(templist)
df.drop(df.iloc[:, 3:12], inplace = True, axis = 1)
df.columns = ['name' ,'todrop', 'value']
df = df.drop(['todrop'], axis = 1)
df['name'].str.strip()
df['value'].str.strip()

def createIndex(df_file):
    df_file['name'].str.strip()
    df_file['value'].str.strip()

    df_file['name'].str.strip()
    df_file['value'].str.strip()

    count = 0 
    name_index = []
    for i in df_file['name']:
        if i == "NAME":
            name_index.append(count)
        count = count + 1 

    count = 0 
    end_index = []
    for i in df_file['name']:
        if i == "END":
            end_index.append(count)
        count = count + 1 
        
    return name_index, end_index

name, end = createIndex(df)

```

```python
def createPD(name_index, end_index, df_file):
    dict_list = []

    for i in range(len(name_index)):
        df = df_file.iloc[name_index[i]:end_index[i]+1,:]
        dict = {}

        for j in range(len(df)):
            dict[df.iloc[j]['name']] = df.iloc[j]['value']

        dict_list.append(dict)
    
    
    return pd.DataFrame(dict_list)

PD = createPD(name,end,df)
```
