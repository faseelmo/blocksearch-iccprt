```python
import numpy as np
import time
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
#pip install ipywidgets

#jupyter nbextension enable --py widgetsnbextension
```

```python
#Importing from csv
start = time.process_time()
df = pd.read_csv('Data/u20.csv', dtype='str')
df = df.drop(df.columns[0], axis = 1)
print('csv time taken:',time.process_time() - start)
```

```python
def search(blockname):
    
    print('searching ...')

    start = time.process_time()

    block = blockname
    
    #THE CODE BELOW RETURNS 1 FOR ALL THE CONTAINS
    
    start = time.process_time()
    
    mask = np.column_stack([df[col].str.contains(block, na=False) for col in df])
    final_pd = df.loc[mask.any(axis=1)]
    x = len(final_pd.index)
    
    
    print('Time taken to mask',time.process_time() - start)
    ##

    start = time.process_time()
    
    
    print(block, 'is used in the following blocks \n')
    for i in range(len(final_pd.index)):

        print("Block:", final_pd.iloc[i][0])
        
        #block_btn = widgets.Button(description=final_pd.iloc[i][1])
        #display(block_btn)

        print("Block Type:", final_pd.iloc[i]['TYPE'])
        print("Description", final_pd.iloc[i]['DESCRP'])

        name = np.where(mask)[1][i]

        print("Input is",final_pd.columns[name] )


        print('')
        
    print('Time taken for looping results ',time.process_time() - start)


  
```

```python
textBox = widgets.Text(description='Block Name' )
display(textBox)

btn = widgets.Button(description='Search')
display(btn)



def btn_eventhandler(obj):
    output_all.clear_output()
    with output_all:
        print('Block name is ', textBox.value)
        search(textBox.value)

    
btn.on_click(btn_eventhandler)


output_all = widgets.Output()

#20LAB10_3TX6:B304.PNT
#Wait for 30 seconds
display(output_all)
```
